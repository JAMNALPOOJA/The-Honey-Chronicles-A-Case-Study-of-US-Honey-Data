# ANALYSIS

# Which states are contributing for most of the years?

```python
df.state.value_counts()
```

```python
plt.figure(figsize= (20,10))
plt.xticks(rotation= 90)

sns.countplot(data = df, x='state')
```

```python
# INFERENCE: States of States Kentucky, New Mexico,Nevada, SouthCarolina, Oklahoma ,Maryland have not contributed in all the 27 years.
```

# Which are the top 5 Honey producing states of all times?

```python
df1 = df.groupby('state').sum()['production'].reset_index()
```

```python
df1
```

```python
df1.sort_values(by='production', ascending= False)
```

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation=90)

sns.barplot(data= df1, x='state', y= 'production')
```

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation=90)

sns.barplot(data= df1.head(), x='state', y= 'production')
```

```python
# Inference:  State NorthDakota, California, SouthDakota, Florida. Montana are the top 5 honey producing states of US from the year 1995 to 2021
```

# What is the change in avg price from 1995 to 2021?

```python
df2 = df.groupby('year').mean().reset_index()
```

```python
df2
```

```python
sns.lineplot(data=df2, x='year', y='average_price')
```

```python
# INFERENCE: From the year 1995 to 2016 honey prices were steadily rising. The prices peaked at 2016 and there was a sudden drop in the prices.
#  and the coming years wittnessed a steady honey price.
```

# Which year witnessed the highest production of Honey in US?

```python
df3 = df.groupby('year').sum().reset_index()
```

```python
df3
```

```python
df3.sort_values(by='production', ascending= False)
```

```python
sns.lineplot(data= df3, x ='year', y='production')
```

```python
# INFERENCE: Honey production peaked in the year 2000 and witnessed a sudden drop in 2010.
```

# In the year 2000, which state contributed the most?

```python
df_2000 = df[df['year']==2000]
```

```python
df_2000.head()
```

```python
df_2000.sort_values(by='production', ascending= False)
```

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation=90)

sns.barplot(data= df_2000, x='state', y='production')
```

```python
# INFERENCE: Top 5 states that contributed in the year 2000 for honey production were NorthDakota, California, SouthDakota, Florida and Minnesota.
```

# In the year 2000, which state had the most colony numbers?

```python
df_2000
```

```python
plt.figure(figsize=(20,10))
plt.xticks(rotation=90)

sns.barplot(data= df_2000.sort_values(by='colonies_number', ascending= False), x='state', y= 'colonies_number')
```

```python
#  Inference: State of California have the highest number of Bee colonies. Even if NorthDakota have highest production
#  still it have second highest number of bee colonies.
```
